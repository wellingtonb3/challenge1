### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


**Analisando o Faturamento da Loja 1**

```python
faturamento1 = loja['Preço'].sum()
print(f"Soma total faturado: R$ {faturamento1:.2f}")
```

**Analisando o Faturamento da Loja 2**

```python
faturamento2 = loja2['Preço'].sum()
print(f"Soma total faturado: R$ {faturamento2:.2f}")
```

**Analisando o Faturamento da Loja 3**


```python
faturamento3 = loja3['Preço'].sum()
print(f"Soma total faturado: R$ {faturamento3:.2f}")
```

**Analisando o Faturamento da Loja 4**

```python
faturamento4 = loja4['Preço'].sum()
print(f"Soma total faturado: R$ {faturamento4:.2f}")
```

# Grafico de Faturamento de Cada Loja

```python
import matplotlib.pyplot as plt

faturamento_total = ["Loja 1", "Loja 2", "Loja 3", "Loja 4"]
valores = [faturamento1, faturamento2, faturamento3, faturamento4]

plt.figure(figsize=(8, 5))
plt.bar(faturamento_total, valores, color=['blue', 'green', 'red', 'purple'])

plt.xlabel("Descrição das Lojas")
plt.ylabel("Valor em R$ Milhões")
plt.title("Comparação dos Faturamentos")

plt.show()
```

# 2. Vendas por Categoria


**Vendas por Categoria na Loja 1**

```python
vendas_por_categoria = loja.groupby('Categoria do Produto')['Preço'].sum().reset_index()
vendas_por_categoria = vendas_por_categoria.sort_values('Preço', ascending=False)

print("Total de vendas por categoria:")
print(vendas_por_categoria)

plt.figure(figsize=(12, 6))
bars = plt.bar(vendas_por_categoria['Categoria do Produto'],
               vendas_por_categoria['Preço'],
               color=plt.cm.tab20.colors)

plt.title('Total de Vendas por Categoria de Produto', fontsize=16, pad=20)
plt.xlabel('Categoria', fontsize=12)
plt.ylabel('Valor Total (R$)', fontsize=12)
plt.xticks(rotation=45, ha='right')

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'R$ {height:,.2f}',
             ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
